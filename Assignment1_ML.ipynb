```python
X=df.drop('Price',axis=1).values
X.shape
```

```python
y=df['Price'].values
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
rms=[]

for i in range(1,10):
  Knn=KNeighborsRegressor(n_neighbors=i)
  Knn.fit(X_train,y_train)
  y_predict=Knn.predict(X_test)


  error=sqrt(mean_squared_error(y_predict,y_test))
  rms.append(error)
  print('for i neigbhors',i,'RMS_Error is',error)

```

```python
#Visulaization of Error Vs Neighbors 
graph=pd.DataFrame(rms)
```

```python
graph.plot()
plt.xlabel('Neighors')
plt.ylabel('RMS_Error')
plt.show()
```

```python
#from the graph predicted the best Neighbor Selection value is n=1 and n=3 because at n=1and n=3 we have the least error.
```

```python
test_Accuracy=Knn.score(X_test,y_test)
train_Accuracy=Knn.score(X_train,y_train)
```

```python
print("Accuracy of the model is")
print("Train Accuracy",train_Accuracy)
print("Test Accuracy",test_Accuracy)
```
